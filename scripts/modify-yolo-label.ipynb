```python
# loop over all txt and modify the labels
import os
dataset_dir = r'C:\Users\dongd\Desktop\pothole_datasets\coco2yolo8.v8i.yolov8'
for part in ['train', 'valid', 'test']:
    label_dir = os.path.join(dataset_dir,part,'labels')
    new_label_dir = label_dir.replace('labels', 'labels_')
    os.makedirs(new_label_dir, exist_ok=True)
    for tf in os.listdir(label_dir):
        tf_path = os.path.join(label_dir,tf)
        with open(tf_path) as file:
            lines = [line.rstrip() for line in file]
        rows = []
        for line in lines:
            row = line.split(' ')
            # modify the below line to change label
            if row[0] == '2': # merge class '2' into '1'
                row[0] = '1' 
            rows.append(row)
        df = pd.DataFrame(rows)
        df.to_csv(os.path.join(new_label_dir,tf), header=None, index=None, sep=' ') 

```
